# Visualization with Matplotlib

Matplotlib is the "grandfather" library of data visualization in Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib((MATLAB-like Plotting Library) will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* High-quality output in many formats
* Very customizable in general
* Great control of every element in a figure but this comes at a cost


Major set back

* Uses more code as compare to other visualization libraries.

Matplotlib allows you to create reproducible figures programmatically. Let's learn how to use it! Before continuing this lecture, I encourage you just to explore the official Matplotlib web page: http://matplotlib.org/

The phases of development in matplotlib are:
1. Prepare data
2. Create plot
3. Customize plot
4. Save plot 
5. Show plot

Import the library

Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

### Changing the Defaults with ``rcParams``

Each time Matplotlib loads, it defines a runtime configuration (rc) containing the default styles for every plot element you create.
This configuration can be adjusted at any time using the ``plt.rc`` convenience routine.
Let's see what it looks like to modify the rc parameters so that our default plot will be of same size as one above.


````python
plt.rcParams['figure.figsize']=(10,6)
````

````python
plt.title ('Loan amount by Sector', fontsize= 16, color='red')
plt.xlabel('Sector')
plt.ylabel('Loan Amount')
plt.xticks(rotation = 90)
plt.ticklabel_format(style='plain')
##plt.yticks(rotation = 60)
plt.plot(sector,loan)
plt.show()
````

*Modify Line Style**

There are two ways of modifying the linestyle. 

The first uses a format string called **fmt** A format string consists of a character for color, marker and line. So to change the marker, linetype and color of a line, you would use something like: **k--x**. This makes line color black, dashed and with x as a marker of the various data points. 

To get what the different possibilities are for **fmt**, check out https://bit.ly/2UXEr6T


````python
plt.title ('Loan amount by Sector', fontsize= 16, color='red')
plt.xlabel('Sector')
plt.ylabel('Loan Amount')
plt.xticks(rotation = 90)
plt.ticklabel_format(style='plain')
##plt.yticks(rotation = 60)
plt.plot(sector,loan,'k--x')
plt.show()
````

The second is a less confusing way of dealing with line types because it involves calling the characteristic directly by use of keyword arguements.


````python
plt.title ('Loan amount by Sector', fontsize= 16, color='red')
plt.xlabel('Sector')
plt.ylabel('Loan Amount')
plt.xticks(rotation = 90)
plt.ticklabel_format(style='plain')
##plt.yticks(rotation = 60)
plt.plot(sector,loan, color='purple', linestyle =':', marker ='o')
plt.show()
````

## Multiline Plot

````python
plt.title ('Loan amount by Sector', fontsize= 16, color='red')
plt.xlabel('Sector')
plt.ylabel('Loan Amount')
plt.xticks(rotation = 90)
plt.ticklabel_format(style='plain')
##plt.yticks(rotation = 60)
plt.plot(sector,loan, color='purple', linestyle =':', marker ='o', label= 'Loan amount')
plt.plot(sector,lender, color = 'black', label= 'Lender Count')
plt.legend()
#plt.grid(True, color = 'red')
plt.show()
````

````python
plt.title('Loan Amount by Sector', fontsize = 16, color = '#662fef', fontstyle = 'oblique')

plt.xlabel('Sector')
plt.ylabel('Loan Amount')
plt.xticks(rotation = 90)
plt.ticklabel_format(style = 'plain')

plt.plot(sector, loan, color = 'cyan', linestyle = ':', marker = 'o', label = 'Loan Amount')
plt.plot(sector, lender, color = 'k', label = 'Lender Count')

plt.legend()
plt.grid(True, color = '#662fef')

plt.show()
````

## Bar Plot

Appropriate for plotting a categorical variable against a numerical variable.  

Use the `plt.bar()` to generate a bar plot.

Syntax: 
```python

```


````python
plt.title('Loan amount by sector')
plt.xlabel('Sector')
plt.ylabel('Loan amount')
plt.xticks(rotation =90)
plt.ticklabel_format(style = 'plain')
plt.bar(sector,loan,color='red', width=0.4) #default width is 0.8
plt.show()
````

## Scatter Plots

You interpret a scatterplot by looking for trends in the data as you go from left to right:

* If the data show an uphill pattern as you move from left to right, this indicates a positive relationship between X and Y. As the X-values increase (move right), the Y-values tend to increase (move up).

* If the data show a downhill pattern as you move from left to right, this indicates a negative relationship between X and Y. As the X-values increase (move right) the Y-values tend to decrease (move down).

* If the data don’t seem to resemble any kind of pattern (even a vague one), then no relationship exists between X and Y.

A linear relationship between X and Y exists when the pattern of X– and Y-values resembles a line, either uphill (with a positive slope) or downhill (with a negative slope).


````python
plt.scatter(loan,lender,marker='x',c='r')
````

## Histogram

By default it returns a plot with 10 bins. Bins are consecutive, non-overlapping intervals of a variable(represented by the bars in a histogram). 

The variable is divided into a series of intervals. A count is then made on how many values fall into each interval.
